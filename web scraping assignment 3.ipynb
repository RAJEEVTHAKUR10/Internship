2. In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
product_name = [] # empty list
Price = [] #empty list

# loop for page iterating
for j in range(3):
    # scraping product name tags
    product_name_tags = driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]')
    for i in product_name_tags:
        if i.text is None :
            product_name.append("-")
        else:
            product_name.append(i.text)

    time.sleep(4)
    # scraping price tags
    price_tags= driver.find_elements_by_xpath('//span[@class="a-price"]')
    for i in price_tags:
        if i.text is None :
            Price.append("-")
        else:
            Price.append(i.text)
    time.sleep(4)
    # page xpath
    next1 = driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()
```

```python
# creating the dataframe from the scraped data
data = pd.DataFrame()
data['Product_name'] = product_name[0:60]
data['Price'] = Price[0:60]
data
```

```python
# saving in csv format
df1 = data.to_csv('guitars.csv')
```

```python
# elements of tags
rank = driver.find_elements_by_xpath('//div[@class="rank"]')
name_tags = driver.find_elements_by_xpath('//div[@class="personName"]')
net_worth_tags = driver.find_elements_by_xpath('//div[@class="netWorth"]')
age_tags = driver.find_elements_by_xpath('//div[@class="age"]')
citizenship_tags =driver.find_elements_by_xpath('//div[@class="countryOfCitizenship"]')
source_tags = driver.find_elements_by_xpath('//div[@class="source-column"]')
industry_tags = driver.find_elements_by_xpath('//div[@class="category"]')
```

```python
# empty list
ranks = []
name = []
net_worth = []
age = []
citizenship = []
source = []
industry = []

#use for loop
for i in rank :
    ranks.append(i.text)
time.sleep(3)
for i in name_tags:
    name.append(i.text)
time.sleep(3)
for i in net_worth_tags:
    net_worth.append(i.text)
time.sleep(3)
for i in age_tags:
    age.append(i.text)
time.sleep(3)
for i in citizenship_tags:
    citizenship.append(i.text)
time.sleep(3)
for i in source_tags:
    source.append(i.text)
time.sleep(3)
for i in industry_tags:
    industry.append(i.text)
time.sleep(3)
```

10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in 
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, 
overall reviews, privates from price, dorms from price, facilities and property description.

```python
url = 'https://www.hostelworld.com/'
driver.get(url)
```

```python
#clicking on search_bar
search_bar1 = driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[2]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/input').click()
time.sleep(3)
# again clicking on search_bar and write london
search_bar = driver.find_element_by_xpath('//input[@id="search-input-field"]')
search_bar.send_keys('London')
time.sleep(3)
```

```python
# select london location
locn = driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[2]/div/div[2]/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]')
locn.click()
time.sleep(3)
```

```python
# clicking the let's go btn
letgo = driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[2]/div/div[2]/div[4]/div/div[2]/div/div[5]/button').click()
time.sleep(6)
```

```python
# element of tags
hotel_name = driver.find_elements_by_xpath('//h2[@class="title title-6"]')
distance = driver.find_elements_by_xpath('//span[@class="description"]')
rating = driver.find_elements_by_xpath('//div[@class="score orange big"]')
total_review = driver.find_elements_by_xpath('//div[@class="reviews"]')
privates_from_price = driver.find_elements_by_xpath('//div[@class="price-col"]')
dorms_price = driver.find_elements_by_xpath('//div[@class="price title-5"]')
facilities = driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')

# empty list
Hotel = []
Distance = []
Rating = []
Total_review = []
privates = []
Dorm = []
Facilities = []

# use for loop
for i in hotel_name:
    Hotel.append(i.text)
time.sleep(2)
for i in distance:
    Distance.append(i.text)
time.sleep(3)
for i in rating:
    Rating.append(i.text)
time.sleep(2)
for i in total_review:
    Total_review.append(i.text)
time.sleep(3)
for i in privates_from_price :
    if i.text is None:
        privates.append("--")
    else:
        privates.append(i.text)
for i in dorms_price:
    if i.text is None :
        Dorm.append("--")
    else:
        Dorm.append(i.text)
for i in facilities:
    Facilities.append(i.text)

# creating the dataframe from the scraped data
data = pd.DataFrame()
data['Hotel_Name'] = Hotel[0:30]
data['Distance from city center'] = Distance [0:30]
data['Total_Review'] = Total_review [0:30]
data['Privates_price'] = privates [0:30]
data['Dorm_price'] = Dorm [0:30]
data['Facilities'] = Facilities [0:30]
data
```
